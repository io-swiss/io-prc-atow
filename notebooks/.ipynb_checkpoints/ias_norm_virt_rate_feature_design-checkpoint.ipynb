```python
import os
import warnings
import matplotlib.pyplot as plt

from tqdm import TqdmExperimentalWarning
from datetime import timedelta
from matplotlib.dates import DateFormatter
from traffic.core import Traffic
import pandas as pd

warnings.filterwarnings("ignore", category=TqdmExperimentalWarning)
warnings.filterwarnings("ignore", category=FutureWarning)

current_directory = os.getcwd()
print("Current Working Directory:", current_directory)
```

## Load data

```python
# Directory containing the Parquet files
directory = '../data/'

traffic_objects = []
for filename in os.listdir(directory):
    if filename.endswith('.parquet'):
        print(f"Loading... {filename}")
        traffic_obj = Traffic.from_file(os.path.join(directory, filename))
        traffic_objects.append(traffic_obj)

# Combine all Traffic objects into a single Traffic object
combined_traffic = sum(traffic_objects)
combined_traffic = combined_traffic.eval()

type(combined_traffic)
```

# Load a single day

```python
t = Traffic.from_file('../data/2022-01-01.parquet')
print(t.data.columns)
flight = t[11]
flight
```

```python
with plt.style.context("traffic"):

    fig, ax = plt.subplots(figsize=(10, 7))

    (
        flight
        .plot_time(
            ax=ax,
            y=["altitude", "vertical_rate"],
            secondary_y=["vertical_rate"]
        )
    )

    ax.set_xlabel("")
    ax.tick_params(axis='x', labelrotation=0)
    ax.xaxis.set_major_formatter(DateFormatter("%H:%M"))
```
